```python
import matplotlib
matplotlib.__version__
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# To create a quick graph using the object-oriented interface method, use the following two lines:
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", label="Boston", color="green", linewidth=2)

# Create labels for the x and y axes.
plt.xlabel("Months")
plt.ylabel("Price($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Use Object oriented approach instead of MATLAB
ax.plot(x_axis, y_axis, color = "green", linewidth=2, marker="*", label="Boston")
ax.set_title("fare by Month")
ax.set_ylim(0, 45)
ax.grid(visible=True, which = "major", axis = "both")
ax.legend()
```

```python
# Create a Vertical Bar Chart
plt.bar(x_axis, y_axis)

# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis)

# to invert y axis in ascending order
plt.gca().invert_yaxis()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
fig, ax = plt.subplots(figsize=(15, 8))
ax.barh(x_axis, y_axis)
```

```python
# Scatter Plot using MATLAB
# Method 1
plt.plot(x_axis, y_axis, 'o')
# Method 2s
plt.scatter(x_axis, y_axis)
```

```python
plt.scatter(x_axis, y_axis, s= y_axis)
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
# Scatter Plot using Object Oriented Interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Scatter Plot using Object Oriented Interface
fif, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Create Pie Charts
plt.pie(y_axis, labels=x_axis)

#add % to each month and explode the largest percentage
explode_values = (0, 0, 0, 0, 0, 0, 0.4, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode = explode_values, labels=x_axis, autopct="%.1f%%")
plt.show()
```
